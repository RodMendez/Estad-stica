```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leer los datos y convertir la columna 'cyl' a factor (categórica)
mtcars = pd.read_csv('smtcars.cv')
mtcars['cyl'] = mtcars['cyl'].astype('category')

# Gráfico de dispersión
sns.scatterplot(data=mtcars, x='wt', y='mpg')
plt.show()
```

```python
# Cambiar el tamaño de los puntos y la forma
sns.scatterplot(data=mtcars, x='wt', y='mpg', size='qsec', sizes=(10, 200), hue='cyl')
plt.show()
```

```python
#Agrega etiquetas del modelo
plt.figure()
for i, row in mtcars.iterrows():
    plt.scatter(row["wt"], row["mpg"], label=row["model"])
plt.legend()
plt.show()
```

```python
#Agregar la línea de regresión con un intervalo de confianza
sns.regplot(data=mtcars, x='wt', y='mpg', scatter_kws={'s': 20})
plt.show()
```

```python
# Eliminar el intervalo de confianza
sns.regplot(data=mtcars, x='wt', y='mpg', scatter_kws={'s': 20}, ci=None)
plt.show()
```

```python
#Regresión lineal con intervalo de confianza al 90%
sns.lmplot(data=mtcars, x='wt', y='mpg', ci=90)
plt.show()
```

```python
# Loess method
sns.lmplot(data=mtcars, x='wt', y='mpg', lowess=True)
plt.show()
```

```python
# Cambiar los colores y las formas de los puntos segun su categoría
sns.scatterplot(data=mtcars, x='wt', y='mpg', hue='cyl', palette='Set2', style='cyl', markers=['o', 's', 'D'])
plt.show()
```

```python
# Cambiar el tipo de línea y el color de la línea de regresión
sns.regplot(data=mtcars, x='wt', y='mpg', scatter_kws={'s': 20}, ci=None, line_kws={'linestyle': '--', 'color': 'red'})
plt.show()
```

```python
# Leer los datos y convertir la columna 'cyl' a factor
mtcars = pd.read_csv('mtcars.csv')
mtcars['cyl'] = mtcars['cyl'].astype('category')

# Cambiar el color de relleno del intervalo de confianza y el estilo de línea
sns.regplot(data=mtcars, x='wt', y='mpg', scatter_kws={'s': 20}, ci=68)
plt.gca().get_lines()[0].set_linestyle("--")
plt.gca().get_lines()[0].set_color("red")
plt.show()
```

```python
# Cambiar las formas y los colores de los puntos manualmente
sns.scatterplot(data=mtcars, x='wt', y='mpg', hue='cyl', style='cyl', markers=['o', 's', 'D'], palette=['#999999', '#E69F00', '#56B4E9'])
plt.show()
```

```python
# Cambiar manualmente el tamaño de los puntos
sns.scatterplot(data=mtcars, x='wt', y='mpg', hue='cyl', style='cyl', size='cyl', sizes=[30, 60, 90], markers=['o', 's', 'D'], palette=['#999999', '#E69F00', '#56B4E9'])
plt.show()
```

```python
# Crear un gráfico de barras
color = ['blue', 'orange', 'purple']
sns.countplot(data=mtcars, x='cyl', palette=color)
plt.title('Número de cilindros')
plt.xlabel('Cilindros')
plt.ylabel('Frecuencias')
plt.show()
```
